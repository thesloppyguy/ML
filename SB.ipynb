```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
cc = sns.load_dataset('car_crashes')
cc
```

```python
# DISTRIBUTION PLOTS
sns.distplot(cc['not_distracted'] ,kde=False,bins= 25)
```

```python
# JOINT PLOT -> COMPARE TWO FEATURES
sns.jointplot(x='speeding',y='alcohol',data=cc,kind='hex')
```

```python
#KDE PLOT 
sns.kdeplot(cc['alcohol'])
```

```python
#PAIR PLOTS -> MAKES EACH POSSIBLE PAIR AND MAKES GRAPHS FOR THEM
sns.pairplot(cc)
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#PAIR PLOT ACC TO A CATEGORICAL DATA
sns.pairplot(data=tips,hue='sex',palette='Blues')
```

```python
# RUG PLOTS -> shit plot
sns.rugplot(tips['tip'])
```

```python
# STYLING
sns.set_style('ticks')
plt.figure(figsize=(8,4))
sns.set_context('paper', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol',data=cc,kind='reg')
```

```python
# CATEGORICAL PLOTS
# BAR PLOT
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.median) # can set estimator your self
```

```python
# COUNT PLOT -> COUNTS CATEGORICAL DATA
sns.countplot(x='sex',data=tips)
```

```python
#BOX PLOT -> FINDS CORE VALUES
sns.boxplot(x='day',y='total_bill',data=tips, hue='sex')
plt.legend(loc=0)
```

```python
#VIOLEN GRAPH
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#STRIP PLOT
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex', jitter=True, dodge=True)
```

```python
#SWARM PLOT
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',split=True,color='white')
```

```python
# MATRIX PLOT
# HEATMAP
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)
cc_mx=cc.corr()
sns.heatmap(data=cc_mx,annot=True,cmap='Blues')
```

```python
# CLUSTER MAP
iris = sns.load_dataset('iris')
iris
```

```python
# HIRARCICAL HEAT MAPS
species = iris.pop('species')
sns.clustermap(iris)
```

```python
# PAIR GIRD
iris = sns.load_dataset('iris')
iris_grid = sns.PairGrid(data=iris,hue="species")
iris_grid.map(plt.scatter)
```

```python
# FACET GRID -> A LOT OF OPTIONS FOR MAKEING IS BUTIFUL
tips_facet = sns.FacetGrid(data=tips,hue='smoker',col='time')
tips_facet.map(plt.scatter,'total_bill','tip')
```
